# 4.新增有效列

1. pandas对DataFrame单列运算，map
2. pandas对DataFrame多列运算，apply

```python
min = df['amnt'].min()
max = df['amnt'].max()
df['col1'] = df['amnt'].map(lambda x: (x-min)/(max-min))
df['col1']
```

```python
df['col2'] = df.apply(lambda x: x['all_prc']-x['net_prc'],axis = 1)
df['col2']
```

```python
import matplotlib.pyplot as plt 
from pylab import mpl

mpl.rcParams['font.sans-serif'] = ['SimHei']
plt.clf()
df[df['date'] == datetime.date(2020, 1, 7)].groupby('bond_tp_nm')['amnt'].sum().plot(kind='bar', figsize=(20, 4))
plt.show()
```

```python
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']

df[df['date'] == datetime.date(2020, 1, 7)].groupby('bond_tp_nm').size().plot(kind = 'bar')
plt.show()
```
